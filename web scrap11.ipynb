```python
from bs4 import BeautifulSoup
import requests
url = BeautifulSoup("https://merolagani.com/LatestMarket.aspx", 'html.parser')
raw_code=requests.get(url)
raw_code
```

```python
soup=BeautifulSoup(raw_code.text,"lxml")
soup
```

```python
raw_head=soup.head
raw_head
```

```python
raw_title=soup.title.string
raw_title
```

```python
raw_title=soup.title
raw_title
```

```python
raw_body=soup.tbody
raw_body
```

```python
data_tags=soup.tbody.find_all("tr")
data_tags
```

```python
tags=soup.tbody.find_all("td")
tags
```

```python
list=[]
for tag in tags:
    x=tag.text
    list.append(x)
list    
    
```

```python
import csv
with open ("merolagani.csv",'w')as file:
    x=csv.writer(file)
    x.writerows(list)
file.close()
```
