```python
from bs4 import BeautifulSoup
import requests
url = BeautifulSoup("https://www.imdb.com/chart/top/", 'html.parser')
raw_context=requests.get(url) 
raw_context                    
                    
```

```python
soup = BeautifulSoup(raw_context.text,"lxml")
soup
```

```python
anchor=soup.find('a')
anchor
```

```python
tags=soup.tbody.find_all('tr')
tags
```

```python
list=[]
for tag in tags:
    x=tag.text.split('\n')
    if x[10]!='':
        
            list.append(x[10:])
list    
    
```

```python
import csv
with open ('flim.csv','w',encoding='utf-8')as file:
    if x!=",":
        
        x=csv.writer(file)
        x.writerows(list)
file.close()    
```
