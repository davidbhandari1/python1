```python
class A:
    def __init__(self,x):
        
        self.x=x
    
class B(A):
    def __init__(self,x,y):
        
        self.y=y
    
        
        A.__init__(self,x)
        
class C(B):
    def __init__(self,x,y,z):
        self.z=z
      
    
        
        B.__init__(self,x,y)
    def area(self):
        area=self.x*self.y
        print("The area is :",area)
    def volume(self):
        volume=self.x*self.y*self.z
        print("The volume is :",volume)
        
            
    



x=int(input("Enter the number for:"))
y=int(input("Enter the number for:"))
z=int(input("Enter the number for:"))
obj=C(x,y,z)
obj.area()
obj.volume()

```

```python
class Length:
    def __init__(self,l):
        self.l = l
        
class Breadth(Length):
    def __init__(self,l,b):
        self.b = b
        Length.__init__(self,l)
    def Area(self):
        area=self.l*self.b
        print("The Area is area:",area)    
       
        
class Data(Breadth):
    def __init__(self,l,b,h):
        self.h = h
        Breadth.__init__(self,l,b)
    
    
    def Volume(self):
        volume=self.l*self.b*self.h
        print("The volume is volume:",volume)
    

l = int(input("Enter length = "))
b = int(input("Enter breadth = "))
h =int(input("Enter height = "))
obj = Data(l,b,h)
obj.Area()
obj.Volume()
```
